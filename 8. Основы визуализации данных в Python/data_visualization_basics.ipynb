## Python для анализа данных - Основы визуализации данных в Python

```python
import numpy as np  # импорт модуля Numpy
import pandas as pd  # импорт модуля Pandas
from matplotlib import pyplot as plt  # импорт инструментов визуализации
from mpl_toolkits.mplot3d import Axes3D  # импорт дополнительной функциональности для построения 3D графиков

# настройка Jupyter Notebook для корректного отображения графиков matplotlib
%matplotlib inline 
```

**Matplotlib** является самым популярным инструментом для визуализации данных Python. Эта библиотека позволяет быстро строить множество различных графиков, подходящих как для анализа так и для публикаций. Ниже мы рассмотрим наиболее распространенные функции данной библиотеки для визуализации данных в интерактивном режиме.

Создадим три массива случайных величин для иллюстрации работы библиотеки matplotlib.

```python
# создание трех одномерноных массивов случайных данных a, b, c
a = np.random.randn(30,)
b = np.random.randn(30,)
c = np.random.randn(30,)
```

### Создание и кастомизация графиков

Наиболее используемая функция matplotlib *plot* по умолчанию при получении одномерного массива данных создает линейный график массива по оси y относительно последовательного индекса по оси x.

```python
plt.plot(a)  # создание графика
plt.show()  # отображение графика
```

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.plot(a)  # создание графика
plt.show()  # отображение графика
```

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.plot(a)  # создание графика
plt.xlabel('x')  # подпись оси x
plt.ylabel('y')  # подбись оси y
plt.show()  # отображение графика
```

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.plot(a)  # создание графика
plt.xlabel('x')  # подпись оси x
plt.ylabel('y')  # подбись оси y
plt.title('Title')  # название графика
plt.show()  # отображение графика
```

```python
plt.figure(figsize=(16,5)) # установка размера рамки графика
plt.plot(a)  # создание графика
plt.xlabel('x')  # подпись оси x
plt.ylabel('y')  # подбись оси y
plt.title('Title')  # название графика
plt.xlim((0,6))  # границы оси x
plt.ylim((-1,1))  # границы оси y
plt.show()  # отображение графика
```

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.plot(a, '--')  # создание графика с явным указанием типа линии --
plt.plot(b)  # создание второго графика
plt.xlabel('x')  # подпись оси x
plt.ylabel('y')  # подбись оси y
plt.title('Title')  # название графика
plt.xlim((0,6))  # границы оси x
plt.ylim((-1,1))  # границы оси y
plt.show()  # отображение графика
```

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.plot(a, '--', label='a')  # создание графика с явным указанием типа линии -- и названия линии для отображения в легенде
plt.plot(b, label='b')  # создание второго графика с явным указанием названия линии для отображения в легенде
plt.xlabel('x')  # подпись оси x
plt.ylabel('y')  # подбись оси y
plt.title('Title')  # название графика
plt.xlim((0,6))  # границы оси x
plt.ylim((-1,1))  # границы оси y
plt.legend()  # создание легенды с названиями элементов графика
plt.show()  # отображение графика
```

```python
# создание диаграммы рассеяния
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.plot(a, b, 'o', label='ab')  # создание диаграммы рассеяния для массивов a и b 
plt.plot(a, c, 'o', label='ac', color='red')  # создание второй диаграммы рассеяния для массивов a и c
plt.xlabel('x')  # подпись оси x
plt.ylabel('y')  # подбись оси y
plt.title('Title')  # название графика
plt.xlim((-1,1))  # границы оси x
plt.ylim((-1,1))  # границы оси y
plt.legend()  # создание легенды с названиями элементов графика
plt.show()  # отображение графика
```

### Создание гистограммы

Для анализа распределения данных часто используется гистограмма. Для создания гистограммы в matplolib используется функция *hist*.

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.hist(a, bins=8)  # создание графика гистограммы с явным указанием количиства делений
plt.title('Histogram')  # название графика
plt.xlabel('Values')  # подпись оси x
plt.ylabel('Frequency')  # подбись оси y
plt.show()  # отображение графика
```

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.hist(a, bins=8, label='histogram of a')  # создание графика гистограммы с явным указанием количиства делений и названием для отображения в легенде
plt.hist(b, bins=8, label='histogram of b')  # создание второго графика гистограммы с явным указанием количиства делений и названием для отображения в легенде
plt.title('Histogram')  # название графика
plt.xlabel('Values')  # подпись оси x
plt.ylabel('Frequency')  # подбись оси y
plt.legend()  # создание легенды с названиями элементов графика
plt.show()  # отображение графика
```

```python
# добавления прозрачности элементам графика с помощью параметра alpha
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.hist(a, bins=8, label='histogram of a', alpha=0.5)  # создание графика гистограммы с явным указанием количиства делений, названием для отображения в легенде и степени програчности
plt.hist(b, bins=8, label='histogram of b', alpha=0.5)  # создание второго графика гистограммы с явным указанием количиства делений и названием для отображения в легенде и степени програчности
plt.title('Histogram')  # название графика
plt.xlabel('Values')  # подпись оси x
plt.ylabel('Frequency')  # подбись оси y
plt.legend()  # создание легенды с названиями элементов графика
plt.show()  # отображение графика
```

### Создание столбчатой диаграммы

Для создания столбчатой диаграммы создадим Pandas DataFrame содержащий 2 колонки со значениями столбцов диаграммы и 1 колонку с названиями столбцов диаграммы. Для создания столбчатой диаграммы в matplotlib используется функция *bar*.

```python
# создание Pandas DataFrame содержащий 2 колонки со значениями столбцов диаграммы и 1 колонку с названиями столбцов диаграммы
df = pd.DataFrame({'a': [10,4,15,11,8],  # значения столбцов диаграммы a
                   'b': [6,13,10,7,12],  # значения столбцов диаграммы b
                   'labels': ['one', 'two', 'three', 'four', 'five']})  # названия столбцов диаграммы
```

```python
df
```

```python
plt.figure(figsize=(16,5))  # установка размера рамки графика
plt.bar(df['labels'], df['a'], alpha=0.5, label='a')  # создание графика столбчатой диаграммы с явным указанием названия для отображения в легенде и степени програчности
plt.bar(df['labels'], df['b'], alpha=0.5, label='b')  # создание второго графика столбчатой диаграммы с явным указанием названия для отображения в легенде и степени програчности
plt.xlabel('labels')  # подпись оси x
plt.ylabel('values')  # подбись оси y
plt.title('Bar Plot')  # название графика
plt.legend()  # создание легенды с названиями элементов графика
plt.show()  # отображение графика
```

### Создание 3D графика

Matplotlib позволяет создавать 3D графики. Для демонстрации функциональности matplotlib по созданию графиков такого типа построим диаграмму рассеяния с 3 осями. Для этого создадим 2 трехмерных массива данных d и e, содержащих по 1000 случайных значений.

```python
d = np.random.randn(1000,3)  # создание трехмерного массива случайных величин
e = np.random.randn(1000,3)  # создание трехмерного массива случайных величин
```

```python
fig = plt.figure(figsize=(18,14))  # установка размера рамки графика
ax = fig.add_subplot(111, projection='3d')  # создание 3D рамки
ax.scatter(d[:,0], d[:,1], d[:,2], marker='o', label='d')  # создание 3D диаграммы рассеяния с явным указанием типа графика и названия графика для отображения в легенде
ax.scatter(e[:,0], e[:,1], e[:,2], marker='o', label='e')  # создание второй 3D диаграммы рассеяния с явным указанием типа графика и названия графика для отображения в легенде
ax.set_xlabel('x', fontsize=20)  # подпись оси x с явным указанием размера шрифта
ax.set_ylabel('y', fontsize=20)  # подпись оси y с явным указанием размера шрифта
ax.set_zlabel('z', fontsize=20)  # подпись оси z с явным указанием размера шрифта
ax.legend()  # создание легенды с названиями элементов графика
plt.show()  # отображение графика
```
